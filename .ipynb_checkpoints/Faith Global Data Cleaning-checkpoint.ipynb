```python
#!pip install openpyxl
import pandas as pd
import numpy as np

# Set maximum rows to display (adjust the number as needed)
pd.set_option('display.max_rows', None)

# Set maximum columns to display (adjust the number as needed)
pd.set_option('display.max_columns', None)
```

```python
# Read Excel data file
df = pd.read_excel("Faith Global Attendance Dataset.xlsx")
df.head(5)
```

```python
# Select columns from index 10 (inclusive) to 25 (exclusive)
cols_to_drop = df.columns[9:26]  

# Drop the selected columns (handles potential mixed data types)
df.drop(cols_to_drop, axis=1, inplace=True)

# Print the modified DataFrame
df.head()
```

```python
# drop extra column not in range of 9:26, copy data 
drop_extra = df.drop(columns = ["Unnamed: 4"])
data = drop_extra.copy()
data.head()
```

```python
# Drop rows with NaN values in 'column_name' column
drop_nans = data[data["Unnamed: 0"].notna()]
drop_nans.head()
```

```python
# Count the occurrences of "Date" (case-insensitive) in 'Unnamed: 0' column
column_values = []
counter = 0

for i in drop_nans["Unnamed: 0"]:
    if "Date" in str(i):  # Check if 'Date' is found (converted to string for case-insensitive matching)
        counter += 1
    column_values.append(counter)
    
# Print the list containing counts
column_values
```

```python
# Add a new column named "Iteration" to 'drop_nans' DataFrame with values from 'column_values'
iter_cols = drop_nans.copy()  # Create a copy to avoid modifying original DataFrame

iter_cols["Iteration"]= column_values

# Print the DataFrame including the new column
iter_cols
```

```python
#Filter for Rows with "Date" (Excluding NaN):
name_df = iter_cols[iter_cols["Unnamed: 0"].notna() & iter_cols["Unnamed: 0"].str.contains("Date")]

name_df.head()
```

```python
# Rename columns of 'name_df' DataFrame with more descriptive names
new_df = name_df.rename(columns = {"Unnamed: 0":"Date", "Unnamed: 1":"Campus", "Unnamed: 2":"Male-Adult",
                          "Unnamed: 3":"Female-Adult",	"Unnamed: 5":"Kids/Teen",	"Unnamed: 6":"Total",
                          "Unnamed: 7":"Global total",	"Unnamed: 8":"Comments"})
new_df.head()
```

```python
# Filter rows where 'Unnamed: 0' does not have a value or does not contain 'Date' (case-insensitive)
noname_df = iter_cols[~(iter_cols["Unnamed: 0"].notna() & iter_cols["Unnamed: 0"].str.contains("Date"))]

# Show the first 5 rows of the new DataFrame
noname_df.head()
```

```python
# Rename columns with more descriptive names and 
clean_data = noname_df.rename(columns = {"Unnamed: 0":"Date", "Unnamed: 1":"Campus", "Unnamed: 2":"Male-Adult",
                          "Unnamed: 3":"Female-Adult",	"Unnamed: 5":"Kids/Teen",	"Unnamed: 6":"Total",
                          "Unnamed: 7":"Global total",	"Unnamed: 8":"Comments"})
#drop 'Iteration' column
clean_data = clean_data.drop(columns = ["Iteration"])

clean_data.head()
```

```python
# Reset index, keeping the old one as a column
clean_data = clean_data.reset_index(drop=False)

# Add a new column with unique IDs starting from 1
clean_data['unique_id'] = clean_data.index + 1 

clean_data.head()
```

```python
clean_data.shape
```

```python
new_clean_data = clean_data.copy()

#  Drops rows with NaN values or values equal to 0.
def cleaned_data(new_clean_data):
  # Drop rows with NaN in specified columns
  new_clean_data = new_clean_data.dropna(subset=['Male-Adult', 'Female-Adult', 'Kids/Teen'])

  # Keep rows where any of the specified columns are not 0
  new_clean_data = new_clean_data[new_clean_data[['Male-Adult', 'Female-Adult', 'Kids/Teen']].any(axis=1)]

  return new_clean_data

cleaned = cleaned_data(new_clean_data.copy())  # Apply logic to a copy

# Print the cleaned DataFrame
cleaned.head()
```

```python
cleaned = cleaned.drop(columns = ["index"])
cleaned.shape
```
